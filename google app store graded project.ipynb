the calculation of the quartiles (q1 and q3),the interquartile range(iqr), and the lower and upper bounds for outlier detection.then filter the rows where the 'Reviews' values fall below the lower bound or above the upper bound to identify the outliers
             </b>

            </span>
        </body>
        </head>

```python
# Check for outliers in the "Reviews" column
Q1 = df['Reviews'].quantile(0.25)
Q3 = df['Reviews'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers = df[(df['Reviews'] > upper_bound) | (df['Reviews'] < lower_bound)]
print("Number of outliers:", len(outliers))
```

```python
#create a boxplot of the 'Reviews'column with automatic outlier detection
sns.boxplot(df['Reviews'])
#set the titles and labels
plt.title('Boxplot of Reviews')
plt.xlabel('Reviews')
#show the plot
plt.show()
```
